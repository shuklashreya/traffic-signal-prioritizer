# Traffic Signal Algorithm for One Side

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from IPython import display
import time
```

```python
imgr=mpimg.imread('R.png')
imgy=mpimg.imread('Y.jpg')
imgg=mpimg.imread('G.jpg')
imgsp=mpimg.imread('SP.jpg')
imggp=mpimg.imread('GP.jpeg')

#method for stop
def stop():
    fig, ax = plt.subplots(1,2)
    ax[0].imshow(imgr)
    ax[1].imshow(imggp)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1.0)

#method for wait
def wait():
    fig, ax = plt.subplots(1,2)
    ax[0].imshow(imgy)
    ax[1].imshow(imgsp)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1.0)

#method for go
def go():
    fig, ax = plt.subplots(1,2)
    ax[0].imshow(imgg)
    ax[1].imshow(imgsp)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1.0)
    
#method for pedestrian check
def pedestrian_check():
    # v = vehicle num
    # p = pedestrian num
    # if v>p:
    # continue
    # if p>v:
    # break and give wait then stop
    # if p==v or v==p:
    # break and give wait then stop
    return None
    
tfinal = time.time() + 15 * 1 #sec * min
while time.time() < tfinal:
    stop()
    br = pedestrian_check()
    #if change in br break and reset loop.
    wait()
    br = pedestrian_check()
    go()
    br = pedestrian_check()
    
```
